```python
from pyspark.sql.functions import current_timestamp, lit

def ingest_to_bronze(
    file_name: str,
    target_name: str
):
    """
    Ingest raw CSV data from Unity Catalog Volume into Bronze Delta table.
    Compatible with Databricks Free Edition (no DBFS usage).
    """

    # 1️⃣ Read RAW CSV from Unity Catalog Volume
    df_raw = (
        spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .option("mode", "PERMISSIVE")
        .csv(f"/Volumes/workspace/default/{PROJECT_NAME}/{file_name}")
    )

    # 2️⃣ Add technical metadata (Bronze layer)
    df_bronze = (
        df_raw
        .withColumn("_ingestion_timestamp", current_timestamp())
        .withColumn("_source_file", lit(file_name))
    )

    # 3️⃣ Write as managed Delta table (Bronze)
    (
        df_bronze
        .write
        .format("delta")
        .mode("overwrite")
        .saveAsTable(f"{BRONZE_SCHEMA}.bronze_{target_name}")
    )

    # 4️⃣ Return ingestion stats
    return {
        "table": f"{BRONZE_SCHEMA}.bronze_{target_name}",
        "rows": df_bronze.count()
    }
```
